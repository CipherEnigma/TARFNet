# Confusion Matrix and Metrics Visualization
This notebook visualizes the confusion matrix from given values and calculates key performance metrics.

```python
import matplotlib.pyplot as plt
import numpy as np


TP = 50
FN = 10
FP = 5
TN = 45

cm = np.array([[TP, FN],
               [FP, TN]])


fig, ax = plt.subplots(figsize=(6, 6))
im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
ax.figure.colorbar(im, ax=ax)


ax.set(xticks=np.arange(cm.shape[1]),
       yticks=np.arange(cm.shape[0]),
       xticklabels=['Predicted Movement', 'Predicted No Movement'],
       yticklabels=['Actual Movement', 'Actual No Movement'],
       ylabel='True label',
       xlabel='Predicted label',
       title='Confusion Matrix')


plt.setp(ax.get_xticklabels(), rotation=20, ha="right", rotation_mode="anchor")


fmt = 'd'
thresh = cm.max() / 2.
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, format(cm[i, j], fmt),
                ha="center", va="center",
                color="white" if cm[i, j] > thresh else "black")

plt.tight_layout()
plt.show()
```

```python

accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * (precision * recall) / (precision + recall)

accuracy, precision, recall, f1_score
```

```python
epochs = list(range(1, 21))
train_acc = [0.46, 0.47, 0.50, 0.50, 0.56, 0.52, 0.56, 0.63, 0.69, 0.60, 0.62, 0.65, 0.57, 0.65, 0.71, 0.66, 0.59, 0.65, 0.78, 0.71]
val_acc   = [0.60, 0.60, 0.67, 0.60, 0.67, 0.80, 0.60, 0.73, 0.67, 0.67, 0.67, 0.60, 0.53, 0.73, 0.60, 0.47, 0.67, 0.60, 0.53, 0.67]

plt.figure(figsize=(8, 5))
plt.plot(epochs, train_acc, marker='o', label='Train Accuracy')
plt.plot(epochs, val_acc, marker='o', label='Validation Accuracy')
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Training vs Validation Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```
